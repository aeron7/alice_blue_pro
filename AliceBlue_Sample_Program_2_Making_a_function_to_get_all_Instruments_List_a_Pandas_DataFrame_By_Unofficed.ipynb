```python
#In Zerodha's API there is a popular function named kite.instruments(). It fetchs all the Instrument token list from Zerodha. Lets try to develop the same for Aliceblue. 
#While looking at the Documentation of API, We can see the link of all Contract Masters. Check here - https://v2api.aliceblueonline.com/contractMaster. Let's use Python's Request to download the contents of the link

import requests
payload = requests.get("https://v2api.aliceblueonline.com/restpy/contract_master?exch=NSE")
```

```python
#Lets check the content of the Payload
payload.text
```

```python
#So this is a JSON Structure. Lets salvage the JSON properly.
payload = payload.json()["NSE"]
payload
```

```python
#The output provides a List of JSON. Now, Let's import the list with the JSON Structures into Pandas Dataframe.
import pandas as pd 

df = pd.DataFrame.from_records(payload)
df
```

```python
#Now lets run a for loop to include the data from all the links of contracts in the ContractMaster page.

df = pd.DataFrame()
for x in ["NSE","NFO","CDS","BSE","BFO","BCD","MCX","INDICES"]:
  payload = requests.get("https://v2api.aliceblueonline.com/restpy/contract_master?exch="+x)
  payload = payload.json()[x]
  df = pd.concat([df,pd.DataFrame.from_records(payload)])

df
```

```python
#So, There is a problem in the "Indices" link. Let's print that separately and checkout the reason. 
requests.get("https://v2api.aliceblueonline.com/restpy/contract_master?exch=INDICES").text
```

```python
#Notice that it is written "BSE". That was obvious too. It is our fault that we assumed there will be some exchange with the name "INDICES". 
#Anyways, We can see that it just contains JSON of symbol of the indices and their token. Like - {"symbol":"SENSEX","token":1}.
#There is no use of this in a trading API. So, Lets just skip it all together and re-write our old function.

df = pd.DataFrame()
for x in ["NSE","NFO","CDS","BSE","BFO","BCD","MCX"]:
  payload = requests.get("https://v2api.aliceblueonline.com/restpy/contract_master?exch="+x)
  payload = payload.json()[x]
  df = pd.concat([df,pd.DataFrame.from_records(payload)])

df
```

```python
#Let's write a small function to get details of a  Trading Symbol. Let's test with Adanient.

instrumentList = df

def get_details(tradesymbol,exchange="NFO"):
    dataToken = instrumentList[(instrumentList['trading_symbol'] == tradesymbol)&(instrumentList['exch']==exchange)]
    return dataToken

tradesymbol = "ADANIENT-EQ"
exchange = "NSE"

get_details(tradesymbol,exchange)
```

```python
#How to get the token of a Trading Symbol? It can be done using one-liner code.

get_details(tradesymbol,exchange).token.iloc[0]
```

```python
#Lets try with a stock options. Lets say AUROPHARMA22SEP540CE

tradesymbol = "AUROPHARMA22SEP540CE"
exchange = "NFO"
get_details(tradesymbol,exchange)
```

```python
#Like token, We can print every other thing which you can see from the column name from the table above.

payload = get_details(tradesymbol,exchange)

#This gives lot size 
print(payload.lot_size.iloc[0])

#This gives lot size 
print(payload.strike_price.iloc[0])

#If you note the expiry date it is coming in some strange format which I suspect to be epoch. Anyways, the point is - We can fetch the value and use it for our trading purposes.
print(payload.expiry_date.iloc[0])
```

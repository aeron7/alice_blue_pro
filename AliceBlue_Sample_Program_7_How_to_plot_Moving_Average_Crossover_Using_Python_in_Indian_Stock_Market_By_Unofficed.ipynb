```python
import datetime
import pandas as pd 
import numpy as np 
from matplotlib import pyplot as plt 

payload = alice.historical_data(alice.get_instrument_by_symbol("NSE", "INFY-EQ"),datetime.datetime(2022, 9, 15),datetime.datetime(2022, 9, 20),HistoricalDataType.Minute)
df = pd.DataFrame.from_records(payload["result"])
df
```

```python
#Let's get the SMA and EMA values from the last exercises. 

MA_period = 20
def sma(price, MA_period):
    mean_price = price.rolling(MA_period).mean()
    return np.round(mean_price, 3)

df['moving_avg'] = sma(df["close"],MA_period)
df['moving_avg_exp'] = df["close"].ewm(span=MA_period,min_periods=MA_period-1).mean()

#Now, The dataframe has values of both Simple Moving Average and Exponential Moving Average. Lets plot a crossover.

df['moving_avg_exp'].plot(c = 'b', figsize = (20, 10), lw = 1)
df['moving_avg'].plot(c = 'g', figsize = (20, 10), lw = 1) 

# show plot
stock_symbol="INFY"
plt.title('Moving Average Crossover: {}'.format(stock_symbol), fontsize = 20)
plt.ylabel('Price in INR(₹)',fontsize = 15 )
plt.xlabel('Date', fontsize = 15 )
plt.legend()
plt.grid()
plt.show()
```
